## Chpater 17

> **Machine learning** is a way to make fast, automatic and good predictions.

> **Prediction** is understood as mapping from a known defined input to an initially unknown, but equally well defined and observable output.

> **Cross validation** is a statistical technique used to evaluate the performance of a machine learning model by partitioning the data into multiple subsets. The most basic cross validation divides the dataset into training set and test set.

> The **training set** is the portion of the data used to fit the model.

> The **test set** is a portion of the data that is hidden from the model during the training phase and it is used to validate the model results.

> A **prediction** is a number that estimates or predicts something. It's the estimation of $\hat{E}[y_i|X_i]$.

> A **policy** is an automatic decision rule. While a prediction is a number, a policy is a decision. \
For example, we can have a policy that engages with customers with income greater than 1000 and doesn't engage otherwise.

> **Target encoding** is a technique used to convert categorical variables into numerical features by replacing each category with the expected value of the target variable for that specific category.\
It is a powerful way to handle categorical variables by leveraging the relationship between the category and the target variable you are trying to predict.

> **Feature engineering** is the process of transforming features to facilitate learning.

> The $R^2$ metric is used to evaluate models that predict a continuous variable. It can go from minus infinity (it will be negative if the prediction is worse than the average) to $1.0$. It tells us how much of the variance in $Y$$ is explained by our model.
